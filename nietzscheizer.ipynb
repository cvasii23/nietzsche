This is a small piece of sowtware that translates day to day english into phrases from Nietzsche's Complete Works.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from bcolz import open as op
import pickle

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity 
from sklearn.manifold import TSNE

from scipy.spatial.distance import cosine
from keras.utils.data_utils import get_file

from IPython.display import YouTubeVideo
```

The input is a given sentence in english. The output will be the most similar phrase in Nietzsche's work. We will have two approaches for the concept of **most similar**:
    * One approach is a TF-IDF one, based of the occurance and importance of the words;
    * A second approach is trying to get to the semantics of the input sentance and is using word embeddings

Nietzsche's Complete works is publicly available on Amazon:

```python
path = get_file('nietzsche.txt', origin="https://s3.amazonaws.com/text-datasets/nietzsche.txt")
text = open(path).read().lower()
print('corpus length:', len(text))
```

We want to split it into sentences, one sentence-one line (not always true, maybe splittind at points would also worth trying):

```python
ntz = text.split('\n')
```

```python
ntz[:5]
```

Now take only letters and remove emply entries:

```python
import re
def onlyletters(s):
    return re.sub('[^a-zA-Z ]+', '', s)
```

```python
ntz = [onlyletters(element) for element in ntz if len(element)>1]
```

```python
ntz[:5]
```

## This is the input sentence in plain, day to day english:

```python
#inputSent = 'i hope this finds you well'
inputSent = 'there is no happiness'
```

Get the TF-IDF most cosine-similar phrase in Nietzsche's work:

## 1. The TF-IDF most similar phrase

```python

def similarities(name):
    documents = [ name ] + ntz
    tf_idf = TfidfVectorizer().fit_transform(documents)
    similaritati  = cosine_similarity(tf_idf[0:1], tf_idf)
    similaritati = list(similaritati[0,1:])
    max_value = max(similaritati)
    max_index = similaritati.index(max_value)
    return(ntz[max_index], max_value)
```

```python
similarities(inputSent)
```

## 2. The most similar phrase by semantics

To get sematic meaning of the phrases, we use word embeddings. Here we have used [GLOVE word embedding from Stanford University](http://nlp.stanford.edu/projects/glove/)

A detail description of what a word embedding is can be found [here, on the Tensorflow page](https://www.tensorflow.org/tutorials/word2vec).

```python
def load_array(fname):
    return bcolz.open(fname)[:]

def load_glove(loc):
    with open(loc+'_words.pkl', 'rb') as f1:
            r1 = pickle.load(f1, encoding='latin1')
    with open(loc+'_idx.pkl', 'rb') as f2:
            r2 = pickle.load(f2, encoding='latin1')
    return (op(loc+'.dat')[:], r1, r2)
```

```python
vecs, words, wordidx = load_glove(r'./6B.300d/6B.300d')
vecs.shape
```

And a function to to the conversion. The first version, the lambda function that I have commented out has not worked only because of Nietzshe that has used some unrealistically rare words:

```python
#w2v = lambda w: vecs[wordidx[w]]
def w2v(w):
    if w in words:
        return vecs[wordidx[w]]
    else:
        return vecs[wordidx['and']]
```

```python
w2v('revenge')[:23]
```

Words with close meaning are mapped onto neaby elements in an $R^n$ space:

```python
tsne = TSNE(n_components=2, random_state=0)
Y = tsne.fit_transform(vecs[:500])

start=0; end=350
dat = Y[start:end]
plt.figure(figsize=(15,15))
plt.scatter(dat[:, 0], dat[:, 1])
for label, x, y in zip(words[start:end], dat[:, 0], dat[:, 1]):
    plt.text(x,y,label, color=np.random.rand(3)*0.7,
                 fontsize=14)
plt.show()
```

This will transform a sentence into the sum of the glove image of its component vector transation of its words. This approach may sound naive, but I decided to sacrifice the injectivity of the translation but preserve the same meaning for two sentences with permuted words.

```python
def phrase2vec(s):
    lista = s.split(' ')
    return sum([w2v(cuv) for cuv in lista])
```

We filter the Nietzsche works by only keepind the sentences with all their words into the 400k vocabulary of our Glove embedding.

```python

ntznew = [s for s in ntz if set(s.split(' ')) < set(words)]
```

```python
ntzVal = [phrase2vec(s) for s in ntznew]
```

```python
inputVal = phrase2vec(inputSent)
```

```python
distCos = [cosine(inputVal, s) for s in ntzVal]
distEucl = [np.linalg.norm(inputVal - s) for s in ntzVal]
```

Euclidean distance is our first candidate, but just for the sake of it, let's also try the cosine:

```python
ntznew[:3]
```

```python
ind1 = np.argmax(distCos)
print( ntznew[ind1])
```

And finally the euclidean distance most similar semantically is:

```python
ind2 = np.argmin(distEucl)
print( ntznew[ind2])
```

## 3. Let's also make a 5 lines summary of all Nietzsche's work

We will perform a k-means clustering. The sentences which are closest to the centroids will provide the summary

```python
from sklearn.cluster import KMeans
```

```python
clustermodel = KMeans(n_clusters=5, init='k-means++',n_jobs=-1)
```

```python

fittedmodel = clustermodel.fit(ntzVal)
centroids = fittedmodel.cluster_centers_
```

```python
indices = []
distance = []
for center in centroids:
    l = []
    l = map(lambda x: np.linalg.norm(x - center), ntzVal)
    val, idx = min((val, idx) for (idx, val) in enumerate(l))
    indices.append(idx)
    distance.append(val)
```

And the summary is:

```python
for indice in indices:
    print(ntznew[indice])
```
